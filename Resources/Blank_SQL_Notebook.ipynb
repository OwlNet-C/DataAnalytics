```python
%%sql

select s.productkey , s.orderdate ,  s.quantity * s.netprice * s.exchangerate as net_revenue,
c.givenname || ' '|| c.surname as fullname, c.countryfull , c.continent , EXTRACT( year from AGE(c.birthday)) as age,
p.productkey , p.productname , p.categoryname , p.subcategoryname
from sales s
left join customer c on c.customerkey = s.customerkey
left join product p on p.productkey = s.productkey
where orderdate::DATE >= '2020-01-01'
limit 5
```

# New section High Value customers

```python
%%sql

select s.productkey , s.orderdate ,  s.quantity * s.netprice * s.exchangerate as net_revenue,
c.givenname || ' '|| c.surname as fullname, c.countryfull , c.continent , EXTRACT( year from AGE(c.birthday)) as age,
p.productkey , p.productname , p.categoryname , p.subcategoryname,
case when (s.quantity * s.netprice * s.exchangerate) > 1000  then 'HighValue' else 'Standard' end as high_standard
from sales s
left join customer c on c.customerkey = s.customerkey
left join product p on p.productkey = s.productkey
where orderdate::DATE >= '2020-01-01'
limit 5
```

# Pivoting Data


```python
%%sql

select orderdate , count( distinct customerkey) as total_customers
from sales
where orderdate between '2023-01-01' and '2023-12-31'
group by orderdate
order by 1
limit 5
```

```python
%%sql

select distinct continent from customer
```

```python
%%sql

select s.orderdate ,
count( distinct s.customerkey) as total_customers,
count(distinct case when c.continent = 'Europe' then s.customerkey end ) as EU_customers,
count(distinct case when c.continent = 'North America' then s.customerkey end ) as NA_customers,
count(distinct case when c.continent = 'Australia' then s.customerkey end ) as AU_customers
from sales s
left join customer c on s.customerkey = c.customerkey
where
  s.orderdate between '2023-01-01' and '2023-12-31'
group by
   s.orderdate
order by 1
```

```python
%%sql

select
  p.categoryname ,
  sum(s.quantity * s.netprice * s.exchangerate) as net_revenue,
  sum(case when s.orderdate between '2022-01-01' and '2022-12-31' then s.quantity * s.netprice * s.exchangerate else 0 end ) as total_net_rev2022,
    sum(case when s.orderdate between '2023-01-01' and '2023-12-31' then s.quantity * s.netprice * s.exchangerate else 0 end ) as total_net_rev2023
from
  sales s
left join product p on s.productkey = p.productkey
group by
   p.categoryname
order by p.categoryname
```

# Statistical Aggregations

```python
%%sql

select
  p.categoryname ,
  avg(case when s.orderdate between '2022-01-01' and '2022-12-31' then s.quantity * s.netprice * s.exchangerate else 0 end ) as Avg_net_rev2022,
  avg(case when s.orderdate between '2023-01-01' and '2023-12-31' then s.quantity * s.netprice * s.exchangerate else 0 end ) as Avg_net_rev2023
from
  sales s
left join product p on s.productkey = p.productkey
group by
   p.categoryname
order by p.categoryname
```
